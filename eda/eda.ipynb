```python
# Import core libraries
import pandas as pd
import numpy as np
```

```python
# Load dataset
df = pd.read_csv("games.csv", encoding='utf-8')

# Display first few rows
df.head()
```

```python
# Shape of the dataset
print("Shape of dataset (rows, columns):", df.shape)

# Overview of datatypes, non-null counts, and memory usage
print("\n--- Dataset Info ---")
df.info()
```

```python
# Missing and non-null values summary
missing_values = df.isnull().sum()
non_null_values = df.notnull().sum()
missing_percentage = (missing_values / len(df)) * 100

missing_summary = pd.DataFrame({
    'Missing Values': missing_values,
    'Non-Null Values': non_null_values,
    'Missing (%)': missing_percentage.round(2)
}).sort_values(by='Missing Values', ascending=False)

print("üîç Missing Value Summary:")
display(missing_summary)
```

```python
# Check for duplicates
duplicate_count = df.duplicated().sum()
print(f"Total Duplicate Rows: {duplicate_count}")

# Count unique values per column
unique_counts = df.nunique().sort_values(ascending=False)
print("\n--- Unique Value Count per Column ---")
display(unique_counts)
```

```python
print("\n--- Columns and Datatypes ---")
df.dtypes
```

```python
# Step 6 (fixed): Parse 'Release date' column safely with flexible formats

# First, look at a few unique date formats
print(df['Release date'].head(10))

# Then convert with dayfirst=True and errors='coerce' to handle mixed formats
df['Release date'] = pd.to_datetime(df['Release date'], format='mixed', errors='coerce', dayfirst=True)

# Verify conversion
print("\nConverted Release date column:")
print(df['Release date'].head(10))

# Check if any nulls were introduced (invalid dates)
print("\nNumber of unparseable dates:", df['Release date'].isna().sum())
```

```python
#Correlation Heatmap
# List of numeric columns we want to analyze correlations for
numeric_cols = [
    'Peak CCU', 'Price', 'User score', 'Positive', 'Negative', 
    'Recommendations', 'Average playtime forever', 'Average playtime two weeks',
    'Median playtime forever', 'Median playtime two weeks', 'Achievements'
]

# Keep only the columns that actually exist in the dataset
num_cols = [col for col in numeric_cols if col in df.columns]

# Proceed only if we have at least 2 numeric columns
if len(num_cols) > 1:
    plt.figure(figsize=(10,6))
    sns.heatmap(df[num_cols].corr(), annot=True, cmap='coolwarm', fmt='.2f')
    plt.title('Correlation Heatmap of Key Numerical Features')
    plt.show()
else:
    print("‚ö†Ô∏è Not enough numeric columns available to plot correlation heatmap.")
```

```python
#Top Games by Positive Reviews Ratio
if 'Positive' in df.columns and 'Negative' in df.columns:
    df['Total Reviews'] = df['Positive'] + df['Negative']
    df['Positive Ratio'] = df['Positive'] / df['Total Reviews']

    top_positive_games = df[['Name','Positive Ratio','Total Reviews']].dropna().sort_values(by='Positive Ratio', ascending=False).head(10)
    print("Top 10 games by positive review ratio:")
    display(top_positive_games)
    
    plt.figure(figsize=(10,5))
    sns.barplot(x='Positive Ratio', y='Name', data=top_positive_games, palette='viridis')
    plt.title('Top 10 Games by Positive Review Ratio')
    plt.xlabel('Positive Review Ratio')
    plt.ylabel('Game Name')
    plt.show()
```

```python
#Popular Genres vs Average User Score
if 'Genres' in df.columns and 'User score' in df.columns:
    genres_df = df[['Genres','User score']].dropna()
    genres_exploded = genres_df.assign(Genres=genres_df['Genres'].str.split(',')).explode('Genres')
    genres_exploded['Genres'] = genres_exploded['Genres'].str.strip()

    top_genre_scores = genres_exploded.groupby('Genres')['User score'].mean().sort_values(ascending=False).head(10)
    
    plt.figure(figsize=(10,5))
    sns.barplot(x=top_genre_scores.values, y=top_genre_scores.index, palette='coolwarm')
    plt.title('Top 10 Genres by Average User Score')
    plt.xlabel('Average User Score')
    plt.ylabel('Genre')
    plt.show()
```

```python
if 'Achievements' in df.columns and 'Average playtime forever' in df.columns:
    plt.figure(figsize=(10,6))
    sns.scatterplot(x='Achievements', y='Average playtime forever', data=df, alpha=0.5)
    plt.title('Achievements vs Average Playtime')
    plt.xlabel('Number of Achievements')
    plt.ylabel('Average Playtime (minutes)')
    plt.ylim(0, df['Average playtime forever'].quantile(0.95))  # avoid extreme tail
    plt.show()
```

```python
#Tag Analysis
if 'Tags' in df.columns:
    tags_series = df['Tags'].dropna().str.split(',').explode().str.strip()
    top_tags = tags_series.value_counts().head(15)
    
    plt.figure(figsize=(10,6))
    sns.barplot(x=top_tags.values, y=top_tags.index, palette='cubehelix')
    plt.title('Top 15 Most Common Game Tags on Steam')
    plt.xlabel('Number of Games')
    plt.ylabel('Tag')
    plt.show()
```

```python
#Price vs Peak Concurrent Users
if 'Price' in df.columns and 'Peak CCU' in df.columns:
    plt.figure(figsize=(10,6))
    sns.scatterplot(x='Price', y='Peak CCU', data=df, alpha=0.6)
    plt.title('Price vs Peak Concurrent Users')
    plt.xlabel('Price ($)')
    plt.ylabel('Peak CCU')
    plt.ylim(0, df['Peak CCU'].quantile(0.95))  # remove extreme outliers
    plt.show()
```

```python
#Free vs Paid Games Analysis
if 'Price' in df.columns and 'Average playtime forever' in df.columns:
    df['IsFree'] = df['Price'] == 0
    
    plt.figure(figsize=(8,5))
    sns.boxplot(x='IsFree', y='Average playtime forever', data=df)
    plt.title('Average Playtime: Free vs Paid Games')
    plt.xlabel('Is Free')
    plt.ylabel('Average Playtime (minutes)')
    plt.ylim(0, df['Average playtime forever'].quantile(0.95))  # remove extreme outliers
    plt.show()
```

```python
#Multiplayer vs SinglePlayer Analysis
if 'Categories' in df.columns and 'Average playtime forever' in df.columns:
    df['IsMultiplayer'] = df['Categories'].fillna('').str.contains('Multiplayer', case=False)
    
    plt.figure(figsize=(8,5))
    sns.boxplot(x='IsMultiplayer', y='Average playtime forever', data=df)
    plt.title('Average Playtime: Multiplayer vs Singleplayer')
    plt.xlabel('Is Multiplayer')
    plt.ylabel('Average Playtime (minutes)')
    plt.ylim(0, df['Average playtime forever'].quantile(0.95))
    plt.show()
```

```python
#Metacritic vs User Score
if 'Metacritic score' in df.columns and 'User score' in df.columns:
    plt.figure(figsize=(8,6))
    sns.scatterplot(x='Metacritic score', y='User score', data=df, alpha=0.6, color='darkorange')
    plt.title("Relationship Between Metacritic and User Scores")
    plt.xlabel("Metacritic Score")
    plt.ylabel("User Score")
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()
```

```python
#Average Playtime By Genre
if 'Average playtime forever' in df.columns and 'Genres' in df.columns:
    plt.figure(figsize=(12,6))
    top_genres = df['Genres'].value_counts().head(10).index
    sns.barplot(data=df[df['Genres'].isin(top_genres)], x='Genres', y='Average playtime forever', estimator=np.mean, palette='coolwarm')
    plt.xticks(rotation=45)
    plt.title("Average Playtime (Forever) by Genre")
    plt.ylabel("Average Playtime (minutes)")
    plt.xlabel("Genre")
    plt.show()
```

```python
#Impact Of Achievements On Gametime
if 'Achievements' in df.columns and 'Average playtime forever' in df.columns:
    df['Achievement_bins'] = pd.cut(df['Achievements'], bins=[0,10,50,100,500,1000,5000], labels=['0-10','11-50','51-100','101-500','501-1000','1000+'])
    playtime_by_ach = df.groupby('Achievement_bins')['Average playtime forever'].mean().dropna()
    plt.figure(figsize=(10,5))
    sns.barplot(x=playtime_by_ach.index, y=playtime_by_ach.values, palette='magma')
    plt.title("Impact of Achievements on Average Playtime")
    plt.xlabel("Achievement Range")
    plt.ylabel("Average Playtime (minutes)")
    plt.show()
```

```python
#Reviews Vs User Score
if {'Positive', 'User score'}.issubset(df.columns):
    plt.figure(figsize=(8,6))
    sns.scatterplot(x='Positive', y='User score', data=df, alpha=0.5, color='purple')
    plt.xscale('log')
    plt.title("Number of Reviews vs User Score")
    plt.xlabel("Number of Positive Reviews (log scale)")
    plt.ylabel("User Score")
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()
```

```python
#Breakdown Of Categories
if 'Categories' in df.columns:
    top_categories = df['Categories'].value_counts().head(15)
    plt.figure(figsize=(10,6))
    sns.barplot(x=top_categories.values, y=top_categories.index, palette='crest')
    plt.title("Top 15 Game Categories on Steam")
    plt.xlabel("Count")
    plt.ylabel("Category")
    plt.show()
```

```python
#Playtime Vs Recommendations
if {'Average playtime forever', 'Recommendations'}.issubset(df.columns):
    plt.figure(figsize=(8,6))
    sns.scatterplot(data=df, x='Average playtime forever', y='Recommendations', alpha=0.5, color='darkcyan')
    plt.xscale('log')
    plt.yscale('log')
    plt.title("Playtime vs Recommendations")
    plt.xlabel("Average Playtime (minutes, log scale)")
    plt.ylabel("Recommendations (log scale)")
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()
```
